<h1> Estudos pandas e plotly </h1>

A biblioteca Plotly oferece uma variedade de opções para criar visualizações interativas. Aqui estão algumas das funções e gráficos mais comuns disponíveis no Plotly Express:

<h2> 1.	Gráficos de Barras: </h2>
    
<p>	px.bar(): Cria gráficos de barras. </p>
<p> px.histogram(): Cria histogramas. </p>

<h2> 2.	Gráficos de Linhas: </h2>

<p> px.line(): Cria gráficos de linhas. </p>
<p> px.scatter(): Cria gráficos de dispersão. </p>
    
<h2>3.	Gráficos de Pizza e Donut: </h2>

<p> px.pie(): Cria gráficos de pizza. </p>
<p> px.sunburst(): Cria gráficos de sunburst. </p>
<p> px.treemap(): Cria gráficos de treemap. </p>

<h2> 4.	Gráficos 3D: </h2>

<p> px.scatter_3d(): Cria gráficos de dispersão tridimensionais. </p>
<p> px.line_3d(): Cria gráficos de linhas tridimensionais. </p>

<h2> 5.	Gráficos Geográficos: </h2>

<p> px.scatter_geo(): Cria gráficos de dispersão geográficos. </p>
<p> px.choropleth(): Cria mapas coropléticos. </p>

<h2> 6.	Gráficos de Caixas: </h2>
	
<p> px.box(): Cria gráficos de caixas. </p>

<h2> 7.	Gráficos de Violino: </h2>

<p> px.violin(): Cria gráficos de violino. </p>

<h2> 8.	Gráficos de Calor: </h2>

<p> px.imshow(): Cria gráficos de calor. </p> 

<h2> 9.	Gráficos Animados: </h2>

<p> px.scatter(), px.line(), etc., com o parâmetro animation_frame: Cria gráficos animados ao longo de uma dimensão. </p>

<h2> 10.	Gráficos de Radar: </h2>

<p> px.line_polar(): Cria gráficos de linhas em coordenadas polares. </p> 

<h2> 11.	Gráficos de Radar (Áreas): </h2>

<p> px.area_polar(): Cria gráficos de áreas em coordenadas polares. </p>

<h2> 12.	Gráficos de Barras Empilhadas ou Agrupadas: </h2>

<p> Configuração do parâmetro barmode para stack ou group em gráficos de barras. </p>
	
<h2> 13 Mapas de Calor (Superfície): </h2>

<p> px.surface(): Cria mapas de calor em 3D </p>
    
<h2> 14. Gráficos de Sankey: </h2>

<p> px.sankey(): Cria gráficos de fluxo de Sankey. </p>

Estes são apenas alguns exemplos, e a Plotly oferece muitas mais opções e personalizações. Recomendo explorar a documentação oficial do Plotly Express para obter detalhes e exemplos mais completos de cada função.

```python
import pandas as pd
import plotly.express as px

# DataFrame de exemplo
data = {'X': [1, 2, 3, 4],
        'Y': [9, 11, 12, 13],
        'Z': [100, 110, 120, 130]}
df = pd.DataFrame(data)

# Gráfico de dispersão 3D
figura = px.scatter_3d(df, x='X', y='Y', z='Z', title='Gráfico de Dispersão 3D', template='plotly_dark')
figura.show()
```

```python
import pandas as pd
import plotly.express as px

# DataFrame de exemplo
data = {'Mês': ['Jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago','set', 'out', 'nov', 'dez'],
        'Faculdade': [1000, 1200, 1500, 1800, 2400, 1200, 300,500, 1560, 2000,456,679],
        'Transporte': [800, 1100, 1400, 1700, 560, 1000, 500,600, 2000, 679, 780, 1300],
        'Alimentacao': [300, 100, 1600, 400, 560, 800, 400, 678, 2000, 549, 1234,457]}
df = pd.DataFrame(data)

# Gráfico de linhas com múltiplas séries
fig = px.line(df, x='Mês', y=['Faculdade', 'Transporte', 'Alimentacao'], 
              title='Custos ao Longo do Tempo', 
              template='plotly_dark')
fig.show()
```

```python
import pandas as pd 
import plotly.express as px

data = {'Cargos': ['Analista', 'cientista de dados', 'programador'], 
        'salario': [3000, 5600, 7000]}

df = pd.DataFrame(data)

figura = px.bar(df, x='Cargos', y= 'salario', 
               title='Gráfico de barras',
               template='plotly_dark')
figura.show()
```

```python
import pandas as pd 
import plotly.express as px 

data = { 'renda_extra' : ['Design', 'freelancer', 'bolo de pote'], 'renda': [2000, 2300, 500]}

df = pd.DataFrame(data) 

grafico = px.funnel(df, x = 'renda_extra', y= 'renda',
                    title='Gráfico funnel 3D',
                    template= 'plotly_dark')

grafico.show()
```
